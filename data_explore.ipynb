```python
num_students = q1q2q3.groupby('class_code').count()[['count']].rename(columns={'count':'num_of_students'})
```

```python
mean_scores = q1q2q3.groupby('class_code').mean().iloc[:, 2:]
```

```python
mean_scores.columns = ['q_11459_mean_score','q_11460_mean_score','q_11461_mean_score']
```

```python
mean_scores = pd.concat([num_students, mean_scores], axis=1)
```

```python
mean_scores
```

```python
mean_scores2 = mean_scores[mean_scores.num_of_students >=5]
```

```python
pos = np.arange(mean_scores2.shape[0])
vals = mean_scores2.iloc[:, 1:].apply(lambda x: (x[0] + x[1] + x[2]) / 3, axis=1).tolist()
```

```python
fig = plt.figure(figsize=(6,8))
plt.barh(pos, vals)
plt.yticks(pos, mean_scores2.index)
plt.title('sum of mean scores by class')
plt.xlabel('scores')
plt.show()
```
